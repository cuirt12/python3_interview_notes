```python
import copy

old_list = [1, 2, 3, 4]
shallow_list = copy.copy(old_list)
deep_list = copy.deepcopy(old_list)

print("=== 1. 看最外层的箱子（列表本身）的地址 ===")
print(f"旧箱子地址:   {id(old_list)}")
print(f"浅拷贝箱子:   {id(shallow_list)}")  # 地址变了！买了个新箱子
print(f"深拷贝箱子:   {id(deep_list)}")     # 地址变了！也买了个新箱子

print("\n=== 2. 看箱子内部的元素（数字 1）的地址 ===")
print(f"旧箱子的元素 1:   {id(old_list[0])}")
print(f"浅拷贝的元素 1:   {id(shallow_list[0])}") # 和旧箱子一模一样！
print(f"深拷贝的元素 1:   {id(deep_list[0])}")    # 竟然也和旧箱子一模一样！
```
